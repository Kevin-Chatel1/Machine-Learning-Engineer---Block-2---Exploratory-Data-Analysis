# Statistiques du dataset

```python
# Importation du dataset
data = pd.read_csv("Speed+Dating+Data.csv", encoding='latin1')
data.head()
```

Chaque participant dans le DataFrame rencontre plusieurs personnes de sexe oppos√© pour chaque vague. Par cons√©quent, afin de compter le nombre r√©el de participants, nous devrions appliquer la m√©thode "drop_duplicates" sur la colonne "iid", qui repr√©sente le num√©ro unique du sujet, du groupe (identifiant de la vague et du sexe)

# 1. Analyse des correspondances lors des rencontres

Nous allons tout d'abord nous int√©resser au nombre total de correspondance de rencontres pour identifier les rdv √† succ√®s, les motifs de rencontres et les scores de comptabilit√© apr√®s un second date.

```python
# D√©fini la taille globale de la figure et des sous-graphes
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Graphique √† secteurs pour les r√©sultats de correspondance
match_counts = data["match"].value_counts()
labels = ["Non", "Oui"]
colors = ['#ff9999','#66b3ff']  
explode = (0.05, 0.05)
axes[0].pie(match_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode)
axes[0].set_title('Pourcentage de correspondance', fontsize=15)

# Graphique √† barres pour l'estimation du score de correspondance
sns.barplot(x=data["match_es"].value_counts().index, y=data["match_es"].value_counts().values, ax=axes[1], palette='viridis')
axes[1].set_title("Distribution des scores de compatibilit√© apr√®s accord pour un second date", fontsize=15)
axes[1].set_xlabel("Score de Correspondance", fontsize=14)
axes[1].set_ylabel("Nombre de participants", fontsize=14)

# D√©fini les x-ticks pour n'afficher que des valeurs enti√®res
axes[1].set_xticks(range(int(data["match_es"].min()), int(data["match_es"].max()) + 1))
axes[1].set_xticklabels([str(i) for i in range(int(data["match_es"].min()), int(data["match_es"].max()) + 1)])

# Am√©liore la mise en page et afficher la figure
plt.tight_layout(pad=2.0)
plt.show()
```

Seulement 15.1%  des participants √† des speed datings parviennent s'engager sur un second rendez-vous.  De plus, parmi ceux qui ont exprim√© un int√©r√™t r√©ciproque, la majorit√© attribue des scores de compatibilit√© mod√©r√©s, d√©montrant que peu de participants sont convaincus de la solidit√© de leur rencontre.  

Cela peut r√©v√®ler un d√©calage entre l'accord pour un second rendez-vous et la confiance r√©elle dans la viabilit√© d'une relation √† long terme sugg√®rant que les participants maintiennent souvent des attentes faiblement optimistes quant aux r√©sultats de leurs rencontres initiales.

```python
# Cr√©ation de la figure avec deux sous-graphiques
fig, axes = plt.subplots(1, 2, figsize=(14, 8))

# D√©fini les objectifs
goals = ["M'amuser", "Rencontrer de nouvelles personnes", "Obtenir un rdv", "Trouver une relation", "Pour l'exp√©rience", "Autre"]

# Configuration des param√®tres pour l'explosion des sections de la pie chart
explode = (0.05, 0.05, 0.05, 0.05, 0.05, 0.05)

# Cr√©ation du graphique √† secteurs pour les objectifs
axes[0].pie(data["goal"].value_counts().values, radius=1.2, autopct='%1.1f%%',
            wedgeprops=dict(width=0.6, edgecolor='w'), startangle=0, pctdistance=0.8, textprops={'fontsize': 16})
axes[0].set_title("Objectifs", pad=40, fontsize=16)
axes[0].legend(goals, bbox_to_anchor=(1.05, -0.05), fontsize=12, ncol=2, shadow=True)

# Cr√©ation du graphique en barres pour les objectifs par genre
sns.countplot(y='goal', data=data, hue='gender', palette=['purple', 'darkcyan'], ax=axes[1])
axes[1].set_yticklabels(goals)
axes[1].set_xlabel("Nombre de personnes", labelpad=12, fontsize=14)
axes[1].set_ylabel("Objectifs", labelpad=12, fontsize=14)
axes[1].legend(bbox_to_anchor=(1, 0.65), fontsize=12, ncol=1, shadow=True)

# Ajustement de la mise en page
plt.tight_layout(pad=2.0)
plt.show()
```

Seulement 13,8% de personnes viennent r√©ellement au rendez-vous dans une optique de chercher un partenaire.

# 4. Analyse de la religion

L'id√©e sera de comprendre les diff√©rentes notations donn√©es par les participants au sujet de la religion et d'identifier si certaines ethnies sont plus sensibles √† ce crit√®re.

```python
# Nombre de vote par notation sur l'importance d'une religion similaire
data["imprelig"].value_counts()
```

```python
# Caract√©risation de l'importance de la religion lors des dates
data_religion = data["imprelig"]
data_religion_count = data_religion.value_counts(ascending=True)
sum_counts = np.sum(data_religion_count.values)
repartition = [np.round(element / sum_counts * 100, 2) for element in data_religion_count.values]

# Cr√©ation du graphique √† barres horizontales
plt.figure(figsize=(12, 8))
bars = plt.barh(data_religion_count.index, repartition, color='skyblue')
plt.title("L'importance d'avoir la m√™me religion chez les deux partenaires", fontsize=15, pad=10)
plt.xlabel("Pourcentage des participants", fontsize=13)
plt.ylabel("Importance", fontsize=13)

# Ajoute des annotations pour les pourcentages
for bar, percentage in zip(bars, repartition):
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2,
             f'{percentage:.1f}%', va='center', fontsize=12, color='black')

plt.xlim(0, 36)
plt.show()
```

La majorit√© des participants ne consid√®rent pas la similitude religieuse comme un crit√®re tr√®s important pour choisir un partenaire dans le cadre du speed dating, accordant des notes mod√©r√©es √† faibles √† cette question.

```python
# Cr√©ation de la figure avec deux sous-graphiques
plt.figure(figsize=(15, 6))

# Premier sous-graphe : Importance de la religion selon le genre
plt.subplot(1, 2, 1)
sns.kdeplot(data=data, x="imprelig", hue="gender", fill=True, common_norm=False, alpha=0.5, 
            linewidth=3, bw_adjust=1.5)
plt.title("Importance d'avoir la m√™me religion selon le genre", pad=12, fontsize=14)
plt.xlabel("Score", labelpad=12, fontsize=20)
plt.ylabel("Densit√©", labelpad=12, fontsize=16)
plt.xlim(0, 10)
plt.ylim(0, 0.2)
plt.legend(title='Genre', loc='upper right', labels=['Femmes', 'Hommes'])

# Deuxi√®me sous-graphe : Importance de la religion selon la race
plt.subplot(1, 2, 2)
sns.kdeplot(data=data, x="imprelig", hue="race", fill=False, common_norm=False, 
            linewidth=3, bw_adjust=1.5, alpha=0.7)
plt.title("Importance d'avoir la m√™me religion selon la race", fontsize=14, pad=10)
plt.xlabel("Score", labelpad=12, fontsize=20)
plt.ylabel("Densit√©", labelpad=12, fontsize=16)
plt.xlim(0, 10)
plt.ylim(0, 0.2)
plt.legend(title='Race', loc='upper right', labels=['Caucasien', 'Asiatique', 'Latino', 'Africain', 'Autre'])

# Titre global de la figure
plt.suptitle("L'importance d'avoir la m√™me religion chez les deux partenaires", fontsize=16)

# Ajustement de la mise en page
plt.tight_layout(pad=1.0)

plt.show()
```

Conform√©ment au graphes pr√©c√©dents, les femmes et les hommes accordent g√©n√©ralement une importance faible √† mod√©r√©e √† la similitude religieuse, avec un pic autour des scores 1-3. Toutefois, les participants africains montrent une densit√© plus √©lev√©e pour les scores √©lev√©s (7-10), indiquant qu'ils attribuent une importance plus grande √† la religion.

Malgr√© tout, ces r√©sultats sugg√®rent que, bien que la religion soit un facteur important pour certains groupes, elle n'est g√©n√©ralement pas cruciale pour la majorit√© des participants.


# 5. L'influence des loisirs dans les dates.

```python
# S√©lectionne des colonnes sp√©cifiques li√©es aux hobbies, au genre et √† l'ethnie des participants
cols = ["gender", "race", "race_o", "sports", "tvsports", "exercise", "dining", "museums", "art", "hiking", 
            "gaming", "clubbing", "reading", "tv", "theater", "movies", "concerts", "music", "shopping", "yoga"]
df_hobby = data[cols]
df_hobby.head(5)
```

```python
# Calculer la moyenne des notes attribu√©es √† chaque hobby par genre
hobby = ["sports", "tvsports", "exercise", "dining", "museums", "art", "hiking", 
            "gaming", "clubbing", "reading", "tv", "theater", "movies", "concerts", "music", "shopping", "yoga"]
            
activities = df_hobby.groupby('gender')[hobby].mean()
activities
```

```python
# Cr√©ation du premier trac√© pour les femmes
trace0 = go.Bar(
    x=activities.columns.values,
    y=activities.iloc[0],        
    name='Femmes',                
    marker=dict(
        color='red'               
    )
)

# Cr√©ation du deuxi√®me trac√© pour les hommes
trace1 = go.Bar(
    x=activities.columns.values,  
    y=activities.iloc[1],         
    name='Hommes',                
    marker=dict(
        color='blue'              
    )
)

# Regroupe les deux trac√©s dans une liste
list_data = [trace0, trace1]

# D√©fini la disposition du graphique
layout = go.Layout(
    title="Int√©r√™t par activit√©s des Hommes vs Femmes",
    font=dict(
        size=16
    ),
    barmode='stack', 
    legend=dict(
        font=dict(
            size=16
        )
    )
)

fig = go.Figure(data=list_data, layout=layout)
fig.show()
```

* Les hommes et les femmes partagent des int√©r√™ts similaires pour des activit√©s comme d√Æner ensemble, regarder des films et la t√©l√©vision, ce qui en fait de bonnes options pour des rendez-vous. 

* Les femmes montrent un int√©r√™t plus √©lev√© pour le shopping, le yoga, la lecture et les mus√©es, tandis que les hommes pr√©f√®rent les sports, l'exercice et les jeux vid√©o.

Par soucis de lisibilit√©, nous allons tout d'abords textualis√© les donn√©es pour une meilleure lisibilit√© avant d'identifier la r√©partition des candidats par typologie de travail et du score d'importance envers ce crit√®re. 

```python
# Renomme les num√©ros par les valeurs textuelles des m√©tiers de la colonne "field_cd"
data['field_cd'] = data['field_cd'].replace({
    1: "Avocat",
    2: "Math",
    3: "Sciences sociales, Psychologue",
    4: "Sciences m√©dicales, pharmaceutiques et biotech",
    5: "Ing√©nierie",
    6: "Anglais/√âcriture Cr√©ative/Journalisme",
    7: "Histoire/Religion/Philosophie",
    8: "Entreprise/Finance",
    9: "√âducation, Acad√©mie",
    10: "Sciences Biologiques/Chimie/Physique",
    11: "Travail Social",
    12: "Undergrad/ind√©cis",
    13: "Science politique/Affaires internationales",
    14: "Cin√©ma",
    15: "Administration des Arts",
    16: "Langues",
    17: "Architecture",
    18: "Autre"
})
```

```python
# Calcul de la r√©partition des m√©tiers
job = data["field_cd"].value_counts(ascending=True)
total = np.sum(job.values)
repartition = [np.round(element/total*100, 2) for element in job.values]

# Cr√©ation du graphique de r√©partition des m√©tiers
plt.figure(figsize=(12, 8))
plt.title("Typologie de travail", pad=10)
plt.barh(job.index, repartition)
plt.xlabel("R√©partition en pourcentage")
for i in range(len(repartition)):
    plt.text(repartition[i], job.index[i], f' {repartition[i]}%', va='center', fontsize=10)

plt.xlim(0, 28)
plt.show()
```

```python
# Regroupement des scores d'affinit√©
def categorize_score(score):
    if score <= 3:
        return 'Faible (0-3)'
    elif score <= 6:
        return 'Moyen (4-6)'
    else:
        return '√âlev√© (7-10)'
```

```python
# Applique la fonction de cat√©gorisation des scores d'affinit√©
data['int_corr_cat'] = data['int_corr'].apply(categorize_score)

# Groupe les donn√©es par domaine d'√©tudes (field_cd) et cat√©gorie de score d'affinit√© (int_corr_cat)
df_job = data.groupby(["field_cd", "int_corr_cat"])["iid"].count().reset_index()
df_job = df_job.rename(columns = {'iid':'count'})

# Cr√©ation du graphique √† barres group√©es
plt.figure(figsize=(14, 10))
sns.barplot(data=df_job, x="count", y="field_cd", hue="int_corr_cat", palette="viridis")
plt.title("R√©partition des participants par domaine d'√©tudes et cat√©gories de score d'affinit√©")
plt.xlabel("Nombre de participants")
plt.ylabel("Domaine d'√©tudes")
plt.legend(title="Cat√©gorie de score d'affinit√©", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Les participants semblent n'avoir que tr√®s peu d'int√©r√™t pour le type de poste occup√©e dans le choix d'un date.

###  E. Corr√©lation des √©valuations moyennes r√©elles VS pr√©f√©rences/croyances/perceptions.

Les donn√©es de √©valuations √©tant su 10 contre 100 pour les autres. Nous allons d'abord remettre √† l'√©chelle les donn√©es avant de concat√©ner les dataframes pour r√©aliser la corr√©lation.

```python
# Ce bloc permet d'identifier les r√©sultats d'√©valuations normalement sur 10, remis √† une √©chelle sur 100 afin de pouvoir identifier les r√©sultats par rapport aux autres graphe. 
evaluation_normalized = evaluation.apply(lambda x: (x / x.sum()) * 100, axis=0)
evaluation_normalized.columns = [f'Eval_Norm_{col}' for col in evaluation_normalized.columns]

# Cette partie fixe les noms des titres de colonnes
preferences.columns = ['Pref_Hommes', 'Pref_Femmes']
perceived.columns = ['Perc_Hommes', 'Perc_Femmes']
beliefs.columns = ['Bel_Hommes', 'Bel_Femmes']
```

```python
# Concat√©nation des dataframes pour faciliter la corr√©lation
combined_data = pd.concat([evaluation_normalized, preferences, perceived, beliefs], axis=1)

# S√©paration des donn√©es pour hommes et femmes
data_hommes = combined_data.filter(like='Hommes')
data_femmes = combined_data.filter(like='Femmes')

# Calcul des matrices de corr√©lation
correlation_matrix_hommes = data_hommes.corr()
correlation_matrix_femmes = data_femmes.corr()

# Visualisation des matrices de corr√©lation
fig, ax = plt.subplots(1, 2, figsize=(24, 10))

sns.heatmap(correlation_matrix_hommes, annot=True, cmap='coolwarm', fmt=".2f", ax=ax[0])
ax[0].set_title('Matrice de corr√©lation pour les Hommes')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=45, horizontalalignment='right')
ax[0].set_yticklabels(ax[0].get_yticklabels(), rotation=0)

sns.heatmap(correlation_matrix_femmes, annot=True, cmap='coolwarm', fmt=".2f", ax=ax[1])
ax[1].set_title('Matrice de corr√©lation pour les Femmes')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=45, horizontalalignment='right')

plt.tight_layout()
plt.show()
```

Les matrices de corr√©lation entre √©valuations r√©elles et perceptions montrent que les hommes adaptent efficacement leur comportement en fonction de ce qu'ils croient que les femmes valorisent, notamment l'attractivit√© et l'intelligence. 

En revanche, les femmes montrent des corr√©lations plus faibles, sugg√©rant une plus grande diversit√© dans les pr√©f√©rences masculines ou une moindre influence de leurs perceptions sur leur comportement. 

Cette analyse r√©v√®le que les hommes ajustent leur comportement pour r√©pondre aux attentes per√ßues, tandis que les femmes adoptent une approche plus authentique lors des rencontres.
